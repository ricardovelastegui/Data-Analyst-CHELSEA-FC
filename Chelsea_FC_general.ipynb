```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from google.colab import drive
drive.mount('/content/drive')
ruta = '/content/drive/MyDrive/estadisticas Chelsea FC.xlsx'
```

```
df = pd.read_excel(ruta,header=None)
```

```
df = df.iloc[2:].reset_index(drop=True)

# Asignar la tercera fila como cabecera (que contiene: Date, Time, Round, etc.)
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)

# Eliminar las columnas con nombre NaN
df = df.loc[:, df.columns.notna()]

# Mostrar resultado limpio
print(df)
```

```
df.head()
```

```
df.info()
```

```
df = df.dropna().reset_index(drop=True)
```

```
df['GF'] = df['GF'].astype('int64')
df['GA'] = df['GA'].astype('int64')
df['xG'] = df['xG'].astype('float64')
df['xGA'] = df['xGA'].astype('float64')
df['Poss'] = df['Poss'].astype('int64')
df['Attendance'] = df['Attendance'].astype('int64')
```

```
df.info()
```

```
cols_cat = ['Venue','Result','Captain','Formation','Opp Formation']
for col in cols_cat:
  print(f'Columna {col}: {df[col].nunique()} subniveles')
```

```
df.describe()
```

```
cols_num = ['GF','GA','xG','xGA','Poss']
fig, ax = plt.subplots(nrows=5,ncols=1,figsize=(8,30))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(cols_num):
  sns.boxplot(x=df[col], ax=ax[i])
  ax[i].set_title(col)
```

```
cols_cat = ['Venue','Result','Captain','Formation','Opp Formation']
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=1)
for i, col in enumerate(cols_cat):
  sns.boxplot(x=df[col], ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=30)
```

```
df.to_excel('data_limpia_chelseaFC.xlsx', index=False)
```

```
ruta_hoja4 = '/content/drive/MyDrive/estadisticas Chelsea FC.xlsx'
df_4 = pd.read_excel(ruta_hoja4, sheet_name='Acciones defensivas',header=None)
df_4.head()
```

```
df_4 = df_4.iloc[2:].reset_index(drop=True)
```

```
df_4.head()
```

```
df_4 = df_4.iloc[:, 1:]
df_4.head()
```

```
df_4.columns = df_4.iloc[0]  # Toma la primera fila como nombre de columnas
df_4 = df_4[1:].reset_index(drop=True)
df_4.head()
```

```
df_4.describe()
```

```
df_4.info()
```

```
df_4 = df_4.dropna()
```

```
df_4['Tkl'] = df_4['Tkl'].astype('float64')
df_4['GF'] = df_4['GF'].astype('float64')
df_4['GA'] = df_4['GA'].astype('float64')

df_4['TklW'] = df_4['TklW'].astype('float64')
df_4['Int'] = df_4['Int'].astype('float64')
df_4['Clr'] = df_4['Clr'].astype('float64')
df_4['Err'] = df_4['Err'].astype('float64')
df_4['Def 3rd'] = df_4['Def 3rd'].astype('int64')
df_4['Mid 3rd'] = df_4['Mid 3rd'].astype('int64')
df_4['Att 3rd'] = df_4['Att 3rd'].astype('int64')
df_4['Att'] = df_4['Att'].astype('int64')
df_4['Tkl%'] = df_4['Tkl%'].astype('float64')
df_4['Lost'] = df_4['Lost'].astype('int64')
df_4['Blocks'] = df_4['Blocks'].astype('int64')
df_4['Sh'] = df_4['Sh'].astype('int64')
df_4['Pass'] = df_4['Pass'].astype('int64')
df_4['Tkl+Int'] = df_4['Tkl+Int'].astype('int64')
```

```
df_4.info()
```

```
df_4.describe()
```

```
errores_promedio = df_4.groupby('Result')['Err'].mean()

print("Errores promedio por tipo de resultado:")
print(errores_promedio)
```

```
despejes_promedio = df_4.groupby('Result')['Clr'].mean()

print("\nDespejes promedio por tipo de resultado:")
print(despejes_promedio)
```
